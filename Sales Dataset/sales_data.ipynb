# Sales Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from pylab import rcParams
rcParams['figure.figsize']=14,6
```

```python
# Load dataset

sales_data=pd.read_csv('sales.csv')
sales_data.head()
```

```python
sales_data.isnull().sum()
```

```python
#Remove missing values

sales_data.dropna(inplace=True)
sales_data.head()
```

```python
# prints information about a DataFrame including the index dtype and columns, non-null values and memory 

sales_data.info()
```

```python
sales_data['Quantity Ordered'].unique()
```

```python
sales_data['Price Each'].unique()
```

```python
sales_data['Quantity Ordered']=pd.to_numeric(sales_data['Quantity Ordered'],errors='coerce')
sales_data['Price Each']=pd.to_numeric(sales_data['Price Each'],errors='coerce')
```

```python
sales_data.isnull().sum()
```

```python
sales_data.dropna(inplace=True)

```

```python
sales_data.info()
```

```python
sales_data.head()
```

```python

sales_data['Month']=sales_data['Order Date'].str[0:2]

sales_data.head()

```

```python
sales_data.info()
```

```python
# cast a pandas object to a specified dtype

sales_data['Month']=sales_data['Month'].astype('int32')
sales_data.head()
```

```python
sales_data['Sale']=sales_data['Quantity Ordered'] * sales_data['Price Each']

sales_data.head()
```

```python
month=sales_data.groupby('Month').sum()

month
```

```python
month.index
```

```python
plt.bar(month.index, month.Sale)
plt.xlabel('Month')
plt.ylabel('Sale')
plt.title('Monthly Sales Analysis')

```

```python
sales_data.head(3)
```

```python
sales_data['Order Date']=pd.to_datetime(sales_data['Order Date'])
sales_data.head(3)
```

```python
sales_data.info()
```

```python
sales_data['Hour']=sales_data['Order Date'].dt.hour

sales_data.head(3)
```

```python
hour=sales_data.groupby('Hour').sum()
hour
```

```python
sales_data.info()
```

```python
plt.bar(hour.index,hour['Quantity Ordered'])
```

```python
plt.plot(hour.index,hour['Quantity Ordered'])
plt.xticks(hour.index)
plt.grid()
plt.show()
```

```python
sales_data['City']=sales_data['Purchase Address'].apply(lambda x:x.split(',')[1]+' '+x.split(',')[2].split(' ')[1])

sales_data.head(3)
```

```python
city=sales_data.groupby('City').sum()
city
```

```python
plt.bar(city.index, city.Sale)
plt.xticks(rotation=90)
plt.show()
```
